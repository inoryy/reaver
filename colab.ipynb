# StarCraft II Deep Reinforcement Learning Agent

## Introduction

This Google Colab Notebook is created as a self-contained example for running StarCraft II DRL training with [Reaver](https://github.com/inoryy/reaver-pysc2) library.  
To use it, simply execute each cell below one by one.Please review Blizzard's terms of the [AI and Machine Learning License](http://blzdistsc2-a.akamaihd.net/AI_AND_MACHINE_LEARNING_LICENSE.html) license beforehand.

## GPU Acceleration

To enable it, go to **Runtime -> Change Runtime Type** and then choose **GPU** from **Hardware acceleration** choice box.

## Training

Now you are all set to start training with Reaver. Execute cell below and look for **RMe** in the log statistics - mean episode rewards.  
In about 30 minutes you should start seeing the agent achieving results comparable to [DeepMind's](https://arxiv.org/abs/1708.04782).

```
!python -m reaver.run --env MoveToBeacon --agent a2c --envs 2 --log_freq 25 --updates 5000 2> stderr.log
```
